```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df["grammar_errors_count"])
plt.title("Grammar errors count")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,4))
plt.subplot(2,2,1)
sns.histplot(df["grammar_errors_count"],bins=15)
plt.title("Grammar errors count")


plt.subplot(2,2,2)
sns.histplot(df["label"], bins=20)
plt.title("Label")


plt.subplot(2,2,3)
sns.histplot(df["source"], bins=[-0.5,0.5,1.5,2.5,3.5,4.5,5.5], shrink=0.8)
plt.title("source")


plt.subplot(2,2,4)
sns.histplot(df["complexity_score"], bins=20)
plt.title("complexity_score")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplot(1,3,1)
sns.countplot(x="sentence_length", data=df)
plt.title("Sentence length")

plt.subplot(1,3,2)
sns.countplot(x="id", data=df)
plt.title("Id")

plt.subplot(1,3,3)
sns.countplot(x="label", data=df)
plt.title("label")

plt.tight_layout()
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
```
