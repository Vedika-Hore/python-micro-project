```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df["age"])
plt.title("Age")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplot(2,2,1)
sns.histplot(df["age"], bins=15)
plt.title("Age")
plt.xlabel("Age")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.histplot(df["bmi"], bins=20)
plt.title("BMI")
plt.xlabel("BMI")
plt.ylabel("Count")


plt.subplot(2,2,3)
sns.histplot(df["children"], bins=[-0.5,0.5,1.5,2.5,3.5,4.5,5.5], shrink=0.8)
plt.title("Children")
plt.xlabel("Children")
plt.ylabel("Count")


plt.subplot(2,2,4)
sns.histplot(df["expenses"], bins=20)
plt.title("Expenses")
plt.xlabel("Expenses")
plt.ylabel("Count")

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplot(1,3,1)
sns.countplot(x="sex", data=df)
plt.title("Sex Count")

plt.subplot(1,3,2)
sns.countplot(x="smoker", data=df)
plt.title("Smoker Count")

plt.subplot(1,3,3)
sns.countplot(x="region", data=df)
plt.title("Region Count")

plt.tight_layout()
```
