```python
import matplotlib.pyplot as plt
column=["price","mileage","year","engineSize","mpg"]

for col in column:
  plt.figure(figsize=(4,4))
  plt.hist(df[col],bins=20,color="skyblue",edgecolor="black")
  plt.title(f'Histogram of {col}')
  plt.xlabel(col)
  plt.ylabel('frequency')
```

```python
# Findings:

# Price:
# Most cars have medium prices, while a few cars have very high prices.

# Mileage:
# Most cars have low to medium mileage. Very high mileage cars are fewer.

# Year:
# Most vehicles are from recent years. Older vehicles are less common.

# Engine Size:
# Most cars have small or medium engine sizes. Large engine cars are rare.

# MPG:
# Most cars have average fuel efficiency. Very high or very low MPG values are uncommon.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
sns.countplot(x=df["fuelType"], edgecolor="black")
plt.title("Fuel Type")
plt.xticks(rotation=90)

plt.subplot(1,3,2)
sns.countplot(x=df["transmission"], edgecolor="black")
plt.title("Transmission")
plt.xticks(rotation=90)

plt.subplot(1,3,3)
sns.countplot(x=df["model"], edgecolor="black")
plt.title("Model")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Findings:

# 1. The category with the highest bar is the most common.
# 2. The category with the lowest bar is the least common.
# 3. Popular fuel types and transmissions have more vehicles.
# 4. Some car models are more common than others.
# 5. These results show which vehicle types are more popular in the market.
```

```python
import seaborn as sns
corr=df.select_dtypes(include=['number']).corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

```python
corr=corr['price'].sort_values(ascending=False)
print("Features are highly correlated with the target variable (price)=",corr)
```

```python
#Observations:

# 1. The heatmap shows the correlation between all numeric features.
# 2. Darker colors indicate stronger positive or negative correlation.
# 3. Features with correlation values closer to 1 are strongly positively correlated with price.
# 4. Features with values closer to -1 are strongly negatively correlated with price.
# 5. Features with values close to 0 have little or no correlation with price.
# 6. The feature(s) with the highest correlation value with price have the strongest influence on car price.
```

```python
#Independent Features (Input Features)
# Model
# Year
# Transmission
# Mileage
# Fuel Type
# Engine Size
# MPG
# Tax

#Dependent Feature (Output/Target Feature)
#price


# Independent Features:
# Model, Year, Transmission, Mileage, Fuel Type, Engine Size, MPG, and Tax
# are independent features because they provide information about the car.

# Dependent Feature:
# Price is the dependent (target) feature because it is the value that
# the model aims to predict based on the input features.

# Reason:
# The characteristics of a car, such as its model, year, mileage,
# engine size, fuel type, and transmission, influence its selling price.
# Therefore, these features are used as inputs to predict the output (Price).
```

```python
from sklearn.preprocessing import StandardScaler
num_cols=["year","price","mileage","tax","mpg","engineSize"]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[num_cols])
scaled_df = pd.DataFrame(scaled_data, columns=num_cols)
print(scaled_df.head())


```

```python
# StandardScaler standardizes the numeric features.
# It transforms the data so that:
# - Mean becomes approximately 0.
# - Standard deviation becomes approximately 1.
# This helps machine learning models perform better because
# all numeric features are on the same scale.
```
